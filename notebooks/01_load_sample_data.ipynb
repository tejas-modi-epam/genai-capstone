```python
# Cell 1: imports
from google.cloud import bigquery
import pandas as pd
import os

project_id = os.getenv("GCP_PROJECT_ID", "your-gcp-project-id")
dataset_id = os.getenv("BQ_DATASET", "cpg_retail")

client = bigquery.Client(project=project_id)
```

```python
# Cell 2: load a CSV to a table
def load_csv_to_bq(csv_path, table_id, schema=None):
    df = pd.read_csv(csv_path)
    job = client.load_table_from_dataframe(df, table_id)
    job.result()
    print(f"Loaded {len(df)} rows into {table_id}")
```

```python
# Cell 3: load all tables
base_path = "../data/raw"

load_csv_to_bq(
    os.path.join(base_path, "dim_store.csv"),
    f"{project_id}.{dataset_id}.dim_store"
)

load_csv_to_bq(
    os.path.join(base_path, "dim_sku.csv"),
    f"{project_id}.{dataset_id}.dim_sku"
)

load_csv_to_bq(
    os.path.join(base_path, "dim_date.csv"),
    f"{project_id}.{dataset_id}.dim_date"
)

load_csv_to_bq(
    os.path.join(base_path, "fact_sales.csv"),
    f"{project_id}.{dataset_id}.fact_sales"
)
```
